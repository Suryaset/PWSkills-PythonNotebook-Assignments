## Ans 1 <br>
Matplotlib is a plotting library for the Python programming language. It provides a wide variety of customizable 2D and 3D plots for visualizing data in different formats.

Matplotlib is used for data visualization and graphical representation of data. It provides an easy-to-use interface for creating interactive visualizations and supports a wide range of formats for saving and exporting figures.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

* Line plot - used to plot a continuous line connecting the data points. 
* Scatter plot - used to plot the data points as individual markers.
* Bar plot - used to plot the data as a series of bars, with the height of each bar representing the value of the data.
* Histogram plot - used to plot the distribution of a dataset by dividing the data into a set of intervals and counting the number of observations in each interval.
* Pie plot - used to plot the data as a circle divided into slices, with the size of each slice representing the proportion of the data.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter plot of x and y")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## Ans 3 
The subplot() function is used to create multiple plots within a single figure. It allows us to arrange multiple plots in a grid format, with each plot representing a different subset of the data or a different visual representation of the same data.

Here's an example code to create four line plots using the subplot() function with the provided data:



```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x3,y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title("Line 4")

plt.show()
```

## Ans 4
A bar plot is a type of plot that represents data using rectangular bars. It is used to compare different categories of data and their corresponding values. Bar plots are commonly used to visualize categorical data and the frequency distribution of categorical variables.

To plot a bar plot and a horizontal bar plot for the given data, we can use the bar() and barh() functions provided by the Matplotlib library.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company,profit)
plt.title("Company, Profit")
plt.xlabel("Company")
plt.ylabel("Profit ($)")
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title("Company Profit")
plt.xlabel("Profit ($)")
plt.ylabel("Company")
plt.show()
```

## Ans 5
A box plot is a graphical representation of a dataset that shows the median, quartiles, and outliers of the data distribution. It is used to identify the spread and skewness of the data and to detect outliers.

To plot a box plot using the given data, we can use the Matplotlib library. Here's the code:

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data for box plots
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating a figure and axis object
fig, ax = plt.subplots()

# Creating box plots for the data
ax.boxplot([box1, box2])

# Setting the title and labels for the plot
ax.set_title("Box Plot of Box1 and Box2")
ax.set_xticklabels(["Box1", "Box2"])
ax.set_ylabel("Values")

# Displaying the plot
plt.show()
```
